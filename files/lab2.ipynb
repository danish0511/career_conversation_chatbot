```python
import os
import json
from dotenv import load_dotenv
from groq import Groq
from openai import OpenAI
import google.generativeai as genai
```

```python
load_dotenv(override=True)
```

```python
groq_api_key = os.getenv("GROQ_API_KEY")
google_api_key =os.getenv("GOOGLE_API_KEY")

if groq_api_key:
    print(f"GROQ API key exists and begins {groq_api_key[:8]}")
else:
    print(f"GROQ API key not set.")
    
if google_api_key:
    print(f"GOOGLE API key exists and begins {google_api_key[:8]}")
else:
    print(f"GOOGLE API key not set.")
```

```python
request = "Please come up with a challenging, nuanced question that I can ask a number of LLMs to evaluate their intelligence. Answer only with the question, no explanation."
messages = [{"role":"user", "content":request}]
```

```python
messages
```

```python
groq = Groq()
response = groq.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=messages
)

question = response.choices[0].message.content
print(question)
```

```python
competitors = []
answers = []
messages = [{"role":"user", "content":question}]
```

```python
model_name = "llama-3.3-70b-versatile"

response = groq.chat.completions.create(
    model = "llama-3.3-70b-versatile",
    messages = messages
)
answer = response.choices[0].message.content
print(answer)

competitors.append(model_name)
answers.append(answer)
```

```python
model_name = "gemini-2.0-flash"

genai.configure(api_key = google_api_key)

gemini = genai.GenerativeModel(model_name)
response = gemini.generate_content(question)
answer = response.text
print(answer)

competitors.append(model_name)
answers.append(answer)
```

```python
# Same format as openai and groq for gemini

gem = OpenAI(api_key = google_api_key, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
model_name = "gemini-2.0-flash"

response = gem.chat.completions.create(
    model=model_name,
    messages=messages
)
answer = response.choices[0].message.content
print(answer)
```

```python
print(competitors)
print(answers)
```

```python
for cometitor, answer in zip(competitors, answers):
    print(f"Competitor: {cometitor}\n\nAnswer: {answer}")
```

```python
together = ""

for index, answer in enumerate(answers):
    together += f"# Response from competitor {index+1}\n\n"
    together += answer + "\n\n"
```

```python
print(together)
```

```python
judge = f"""You are  judging a competition between {len(competitors)} competitors.
Each model has been given this question:
{question}

Your job is to evaluate each response for clarity and strength of arguements, and rank them in order of best to worst. Respond with JSON and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", ...]}}

Here are the responses from each competitor:
{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks.
"""
```

```python
print(judge)
```

```python
judge_messages = [{"role": "user", "content": judge}]
```

```python
groq = Groq()
response = groq.chat.completions.create(
    model = "llama-3.3-70b-versatile",
    messages = judge_messages
)
results = response.choices[0].message.content
print(results)
```

```python
result_dict = json.loads(results)
ranks = result_dict["results"]
for index, result in enumerate(ranks):
    competitor = competitors[int(result)-1]
    print(f"Rank {index+1}: {competitor}")
```
